Load in the NP Files

```python
data = np.load('./data/convert/train/RI02.npz')
data.keys
```

```python
spectrograms = data['arr_0']
coords = data['arr_1']  
```

Visualize Spectrogram Function

Left/Right Comparison

```python
def visualize_spect(i):
    curr_spect = spectrograms[i]
    _, axs = plt.subplots(nrows = 1, ncols=2, figsize=(12, 4))
    axs[0].imshow(curr_spect[0], origin='lower', aspect='auto', cmap='inferno')
    axs[0].set_title('Left Channel')
    axs[0].set_xlabel('Time')
    axs[0].set_ylabel('Frequency')
    axs[0].set_ylim(0,100)
    axs[1].imshow(curr_spect[1], origin='lower', aspect='auto', cmap='inferno')
    axs[1].set_title('Right Channel')
    axs[1].set_xlabel('Time (ms)')
    axs[1].set_ylabel('Frequency')
    axs[1].set_ylim(0,100)
    plt.suptitle(f"Spectrogram for Sample #{i}\nAzimuth: {np.degrees(np.arctan2(coords[i,1], coords[i,0])):.1f}Â°")
    plt.tight_layout()
    plt.show()  
```

```python
visualize_spect(2000)
```
